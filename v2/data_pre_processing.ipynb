```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) >= threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

- Entropia (Shannon): Mede a incerteza ou aleatoriedade dos dados. Para dados contínuos, podemos discretizá-los ou usar estimativas de densidade.
- Complexidade: Pode ser medida de diversas formas. Uma abordagem comum é utilizar a Complexidade de Lempel-Ziv ou métricas baseadas em séries temporais.

```python
import numpy as np
import pandas as pd
from scipy.stats import entropy
from sklearn.preprocessing import KBinsDiscretizer

# Função para calcular a entropia
def calculate_entropy(series, bins=10):
    discretizer = KBinsDiscretizer(n_bins=bins, encode='ordinal', strategy='uniform')
    discretized = discretizer.fit_transform(series.values.reshape(-1, 1)).flatten()
    counts, _ = np.histogram(discretized, bins=bins, range=(0, bins))
    return entropy(counts, base=2)

# Função de exemplo para calcular a complexidade (usando a Complexidade de Lempel-Ziv)
def calculate_complexity(series):
    binary_str = ''.join(['1' if x > series.median() else '0' for x in series])
    complexity = 0
    i = 0
    n = len(binary_str)
    k = 1
    while i + k <= n:
        if binary_str[i:i+k] in binary_str[:i]:
            k += 1
            complexity += 1
        else:
            i += k
            k = 1
    return complexity

numeric_cols = final_df.select_dtypes(include=[np.number]).columns.tolist()
print(numeric_cols)
print(len(numeric_cols))
numeric_cols.remove('target') 

for col in numeric_cols:
    entropy_col = f"{col}_entropy"
    complexity_col = f"{col}_complexity"
    
    # Initialize columns with float type by using 0.0
    final_df[entropy_col] = 0.0
    final_df[complexity_col] = 0.0
    
    for target in final_df['target'].unique():
        mask = final_df['target'] == target
        entropy_value = calculate_entropy(final_df.loc[mask, col])
        complexity_value = calculate_complexity(final_df.loc[mask, col])
        
        final_df.loc[mask, entropy_col] = entropy_value
        final_df.loc[mask, complexity_col] = complexity_value

final_df.to_csv('final_df.csv', index=False)
```
